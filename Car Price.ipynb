# Data Preparation

```python
df['CarName'] = df['CarName'].str.split(' ',1).str[0]  #Removing everything except car's company name
df.CarName.value_counts()  #Checking the entries
```

```python
#Correcting the typos
df['CarName'] = df['CarName'].replace('toyouta','toyota')
df['CarName'] = df['CarName'].replace('Nissan','nissan')
df['CarName'] = df['CarName'].replace('vokswagen','volkswagen')
df['CarName'] = df['CarName'].replace('vw','volkswagen')
df['CarName'] = df['CarName'].replace('porcshce','porsche')
df['CarName'] = df['CarName'].replace('maxda','mazda')
df.CarName.value_counts()
```

```python
#Selecting the categorical variables which only have two outcomes like yes/no.
varlist = ['fueltype','aspiration','doornumber','enginelocation']

#Function to map the outcome to either 0 or 1
def binary_map(x):
    return x.map({'gas':0,'diesel':1,'std':0,'turbo':1,'two':0,'four':1,'front':0,'rear':1})

#Applying the function to selected variable list
df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head()  #Checking the first five rows of the dataset
```

Now we convert the remaining categorical variables with more than two levels to a numeric variable

```python
#Converting the categorical variable and storing it in new variable - bodytype while dropping the first column
bodytype = pd.get_dummies(df['carbody'], drop_first=True)
```

```python
#Checking the converted data
bodytype.head()
```

```python
drive = pd.get_dummies(df['drivewheel'], drop_first=True)
```

```python
engine = pd.get_dummies(df['enginetype'], drop_first=True)
```

```python
cylinder = pd.get_dummies(df['cylindernumber'], drop_first=True)
```

```python
fuel = pd.get_dummies(df['fuelsystem'], drop_first=True)
```

```python
car = pd.get_dummies(df['CarName'], drop_first=True)
```

```python
#Merging the converted categorical variables with the df dataframe
df = pd.concat([df,car,bodytype,drive,engine,cylinder,fuel], axis=1)
```

```python
df.drop(['CarName','carbody','drivewheel','enginetype','cylindernumber','fuelsystem'], axis=1, inplace=True)
```

```python
df.info()
```

## Model

### RFE - First Model

```python
#importing RFE and linear regression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Running RFE with a total of 10 variables
lr = LinearRegression()
lr.fit(X_train,y_train)

#Running RFE
rfe = RFE(lr,10)
rfe.fit(X_train,y_train)
```

```python
#Checking the ranking and inclusion of all variables
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Checking the non-included variables
X_train.columns[~rfe.support_]
```

## Building First model using statsmodel for detailed statistics

```python
#Creaing X_train dataframe using the variables selected in RFE model
X_train_rfe = X_train[col]
```

```python
#Importing library & Adding a constant
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lr = sm.OLS(y_train,X_train_rfe).fit() #Running the linear model
```

```python
lr.summary() #Checking the statistic summary
```

#### Variable BMW and Porsche is not a feature but a car brand which just states that performance vehicles are highly sought after in US market. So we can drop these two variables - bmw and porsche at once.

#### Also the same rule applies for the rest of the variables which reflects the car brand. A car brand is just a symbolism of kind of cars people expect from the manufacturer. For example bmw, porsche, audi, jaguar are some of the manufacturer who makes performance oriented vehicles whereas toyota, volkswagen, nissan, volvo etc are generic car manufacturers who makes all-rounder type of cars which are good in all aspects but are in no sense a performance vehicle.

```python
#Dropping the car brands
X_train.drop(['audi','bmw','chevrolet','dodge','honda','isuzu','jaguar','mazda','buick','mercury','mitsubishi','peugeot','plymouth','porsche','renault','saab','subaru','toyota','volkswagen','volvo'],axis=1, inplace=True)
```

### Checking VIF

```python
#Importing library for checking VIFs
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Creating a dataframe to store variable names used in the model and their respective VIF values
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values,i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending=False)
vif
```

#### Post running VIF on second model we see that enginesize and curbweight are high VIFs. However rotor and two cylinder variables have infinite VIFs

```python
#Dropping horsepower variable as p-value should be taken into account first
X_train_rfe.drop(['horsepower'],axis=1,inplace=True)
```
